```python
import pandas as pd
from googletrans import Translator
import locale
base= pd.read_csv('train.csv')



# Função para traduzir um texto
def traduzir_texto(texto):
    translator = Translator()
    try:
        return translator.translate(texto, dest='pt').text
    except:
        return texto

# Traduzir os nomes das colunas
base.columns = [traduzir_texto(coluna) for coluna in base.columns]

# Salvar o DataFrame com os novos nomes das colunas
base.to_csv('seuarquivo_traduzido.csv', index=False)
```

```python
import pandas as pd

# Supondo que 'base' seja o seu DataFrame
# Se precisar, substitua 'base' pelo nome do seu DataFrame

# Defina uma função para formatar os valores em reais abreviados
def formatar_em_reais(valor):
    return f'R$ {valor:,.2f}'.replace('.', ',')

# Encontrar o maior valor do empréstimo
maior_valor_emprestimo = base['Montante do empréstimo'].max()

# Encontrar o menor valor do empréstimo
menor_valor_emprestimo = base['Montante do empréstimo'].min()

# Encontrar a média do valor do empréstimo
media_valor_emprestimo = base['Montante do empréstimo'].mean()

# Calcular a soma dos valores do empréstimo
soma_valor_emprestimo = base['Montante do empréstimo'].sum()

# Formatar os resultados em reais abreviados
maior_valor_emprestimo_formatado = formatar_em_reais(maior_valor_emprestimo)
menor_valor_emprestimo_formatado = formatar_em_reais(menor_valor_emprestimo)
media_valor_emprestimo_formatado = formatar_em_reais(media_valor_emprestimo)
soma_valor_emprestimo_formatado = formatar_em_reais(soma_valor_emprestimo)

# Exibir os resultados formatados
print(f"Maior Valor do Empréstimo: {maior_valor_emprestimo_formatado}")
print(f"Menor Valor do Empréstimo: {menor_valor_emprestimo_formatado}")
print(f"Média do Valor do Empréstimo: {media_valor_emprestimo_formatado}")
print(f"Soma do Valor do Empréstimo: {soma_valor_emprestimo_formatado}")
```
